#### 1. Import pandas library

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

#### 2. Import users table:

```python
users_df = pd.read_csv('users_table.csv')
users_df.head()
```

#### 3. Rename Id column to userId

```python
users_df = users_df.rename(columns={"Id": "userId"})
users_df.head()
```

#### 4. Import posts table:

```python
posts_df = pd.read_csv('posts_table.csv')
posts_df.sample(10)
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts_df = posts_df.rename(columns={"Id": "postId", "OwnerUserId": "userId"})
posts_df.sample(10)
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
users_columns = users_df[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
posts_columns = posts_df[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
```

#### 7. Merge both dataframes, users and posts.
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_dfs = posts_columns.merge(users_columns, on='userId')
merged_dfs.sample(10)
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain.
**Remember** to check the results of your code before passing to the next step

```python
"""The field viewCount seems to be the quantity of visualizations in a post.
Therefore, no vizualization is triggering null instead of zero.
However, 35% of the entries have null amounts for viewCounts and values greater
than zero for CommentCount, which does not make sense, as the user must
view the post to be able to comment. Unless the CommentCount is related
to a different thing"""

merged_dfs['ViewCount'].fillna(0, inplace=True)
merged_dfs.sample(10)
```
