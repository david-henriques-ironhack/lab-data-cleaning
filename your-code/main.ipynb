#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
user_table = pd.read_csv('datasets/users_table.csv')

user_table.head()
```

#### 3. Rename Id column to userId

```python
user_table = user_table.rename(columns={'Id' : 'userId'})
user_table.head()
```

#### 4. Import posts table:

```python
posts_table = pd.read_csv('datasets/posts_table.csv')

posts_table.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts_table = posts_table.rename(columns={'Id' : 'postId', 'OwnerUserId' : 'userId'})

posts_table.head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
users_columns = user_table[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]

posts_columns = posts_table[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
final_df = pd.merge(user_table, posts_columns)

final_df.tail()
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
final_df.isna().sum()
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
# the total number of rows is 38962, this leads that missing values over 15000 can be removed in order to have more constant data

columns_to_remove= final_df.columns[final_df.isna().sum() > 15000]
final_df_cleaned = final_df.drop(columns=columns_to_remove)

final_df_cleaned.isnull().sum()

# now we have the DisplayName column with only 10 missing values, and we can replace them with a common value

final_df_cleaned['DisplayName'].value_counts()

final_df_cleaned[final_df_cleaned.isnull()].head()

# checking these 10 null values shows that all rows contain a lot of no value, so you can delete these 10 rows

condition = final_df_cleaned[final_df_cleaned['DisplayName'].isnull()]

index_to_remove = condition.index[:10]

final_df_cleaned = final_df_cleaned.drop(index=index_to_remove)

print(final_df_cleaned.isnull().sum())

final_df_cleaned.head()
```
