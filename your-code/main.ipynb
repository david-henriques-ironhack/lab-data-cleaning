#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
posts_selected = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
print(posts_selected.head())
```

```python
users_selected = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
print(users_selected.head())
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_df = pd.merge(posts_selected, users_selected, on='userId', how='inner')
print(merged_df.head())
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
missing_values = merged_df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
merged_df_cleaned = merged_df.dropna()
missing_values_after = merged_df.isnull().sum()
print(missing_values_after[missing_values_after > 0])
```
