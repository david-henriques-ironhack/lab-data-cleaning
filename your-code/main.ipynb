#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import users table:

```python
import os
import rarfile
```

```python
# Get the current working directory
current_dir = os.getcwd()
```

```python
# Path to the rar file 
rar_file_path = os.path.join(current_dir, 'datasets.rar')
```

```python
# Set the extraction path 
extraction_path = os.path.join(current_dir, 'extracted_files')
os.makedirs(extraction_path, exist_ok=True)  # Create the extraction directory 
```

```python
# Extract the rar file
with rarfile.RarFile(rar_file_path) as rf:
    rf.extractall(path=extraction_path)
```

```python
users_df = pd.read_csv(f"{extraction_path}/users_table.csv")
```

```python
users_df.head()
```

#### 3. Rename Id column to userId

```python
users_df.rename(columns = {'Id':'userId'}, inplace = True)
users_df.head()
```

#### 4. Import posts table:

```python
posts_df = pd.read_csv(f"{extraction_path}/posts_table.csv")
```

```python
posts_df.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts_df.rename(columns={'Id':'postId','OwnerUserId':'userId'}, inplace=True)
posts_df.head()
```

```python
posts_df.columns
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
new_users_df = users_df[['userId','Reputation','Views','UpVotes','DownVotes']]
new_posts_df = posts_df[['postId','Score','userId','ViewCount','CommentCount']]
```

```python
new_users_df
```

```python
new_posts_df
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_df = pd.merge(new_users_df,new_posts_df, on='userId', how='outer')
merged_df
```

```python
merged_df.columns
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
# Drop rows with missing 'userId' or 'postId' as these are key identifiers which would mean they have a unique identifier and it's aprox 1% of the data it won't have a great impact
merged_df = merged_df.dropna(subset = ['userId','postId'])
merged_df.shape
```

```python
# Fill missing values with the median for 'Reputation', 'Score', 'ViewCount', 'CommentCount' is a more balanced option as it helps maintain the central distribution of the data without introducing extreme values like 0
median_columns = ['Reputation', 'Score', 'ViewCount', 'CommentCount']
merged_df[median_columns] = merged_df[median_columns].fillna(merged_df[median_columns].median())
merged_df
 
```

```python
# Fill missing values with 0 for 'Views', 'UpVotes', 'DownVotes' where it likely represents no activity
zero_columns = ['Views', 'UpVotes', 'DownVotes']
merged_df[zero_columns] = merged_df[zero_columns].fillna(0)
merged_df
```

```python
merged_df.info()
```

#### 10. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
# Convert 'userId' and 'postId' to integers to optimise memory
merged_df['userId'] = merged_df['userId'].astype('int64')
merged_df['postId'] = merged_df['postId'].astype('int64')
```

```python
# Convert other numeric columns to integers to optimise memory
numeric_columns = ['Reputation', 'Views', 'UpVotes', 'DownVotes', 'Score', 'ViewCount', 'CommentCount']
merged_df[numeric_columns] = merged_df[numeric_columns].astype('int64')
```
